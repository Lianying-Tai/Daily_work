```python
from matplotlib import pyplot as plt
x =[1,2,3,4,5]
y = [2,3,4,5,6]
plt.plot(x,y)
```

```python
fig1 = plt.figure(figsize=(5,5))
plt.plot(x,y)
plt.title('y vs x')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
fig2 = plt.figure(figsize=(5,5))
plt.scatter(x,y)
plt.title('y vs x')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import pandas as pd
data = pd.read_csv('data.csv')
print(type(data))
print(data)
```
